### Handling outliers

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#outliers
plt.boxplot(data['no_of_trainings'])
```

```python
sns.boxplot(data['age'])
```

```python
#Finding the different quartile values
Q1 = np.percentile(data['age'],25,interpolation='midpoint')
Q2 = np.percentile(data['age'],50,interpolation='midpoint')
Q3 = np.percentile(data['age'],75,interpolation='midpoint')
print(Q1)
print(Q2)
print(Q3)
```

```python
IQR = Q3-Q1 ; print("IQR is",IQR)
low_limit = Q1-1.5*IQR ; print("lower limit is",low_limit)
upper_limit = Q3+1.5*IQR ; print("upper limit is",upper_limit)
```

```python
outlier=[]
for x in data['age']:
    if((x > upper_limit)or(x < low_limit)):
        outlier.append(x)
```

```python
ind1 = data['age']>upper_limit
index = data.loc[ind1].index
data.drop(index,inplace=True)
sns.boxplot(data['age'])
plt.title('Box plot of age')  
```

## Modelling

```python
## Logistic regression model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2,test_size=0.3)

from sklearn.linear_model import LogisticRegression
```

```python
x_train.shape
```

```python
logit_model=LogisticRegression()
logit_model.fit(x_train,y_train)
y_pred=logit_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score

print("Accuracy is:",accuracy_score(y_test,y_pred))
print("Precision is:",precision_score(y_test,y_pred))
print("Recall is:",recall_score(y_test,y_pred))
print("F1 score is:",f1_score(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
##k-NN

from sklearn.neighbors import KNeighborsClassifier

acc_values = []
neighbors = np.arange(3,15)
for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors=k,metric='minkowski')
    classifier.fit(x_train,y_train)
    y_pred = classifier.predict(x_test)
    acc = accuracy_score(y_test,y_pred)
    acc_values.append(acc)
```

```python
plt.plot(neighbors, acc_values,'0.')
plt.xlabel('k value')
plt.ylabel('accuracy')
plt.show()
```

```python
classifier = KNeighborsClassifier(n_neighbors=5,metric='minkowski')
classifier.fit(x_train,y_train)
y_pred = classifier.predict(x_test)
```

```python
print("Accuracy is:",accuracy_score(y_test,y_pred))
print("Precision is:",precision_score(y_test,y_pred))
print("Recall is:",recall_score(y_test,y_pred))
print("F1 score is:",f1_score(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
